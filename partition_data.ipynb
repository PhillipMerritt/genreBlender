```python
import os
import shutil
from collections import defaultdict
import re
import random
from tqdm import tqdm
random.seed(1)

#artist_counts = defaultdict(int)
songs = defaultdict(list)
parent_folder = "data/130000_Pop_Rock_Classical_Videogame_EDM_MIDI_Archive[6_19_15]/Classical_Guitar_classicalguitarmidi.com_MIDIRip/"
filenames = os.listdir(parent_folder)
#artist_names = []

for f in filenames:
    match = re.search(r"^([a-zA-Z0-9]+)_", f)
    if match:
        songs[match.group(1)].append(f)

count = 0.0
mini = 99
maxi = 0

ignored = set()
artists = []
counts = []

for key, value in songs.items():
    c = len(value)
    counts.append(c)
    if c < 5:
        ignored.add(key)
    else:
        artists.append(key)
        count += c
        mini = min(mini, c)
        maxi = max(maxi, c)

print("total songs: {}".format(int(count)))

mean = int(count / len(artists))
print(mean)
print(maxi)
print(mini)

counts = sorted(counts)
print(counts)

size = 30

random.shuffle(artists)
train_artists = artists[:int(len(artists) * .8)]
val_artists = artists[int(len(artists) * .8) : int(len(artists) * .9)]
test_artists = artists[int(len(artists) * .9):]

train_files = []
val_files = []
test_files = []

for a in train_artists:
    arr = songs[a]

    if len(arr) > size:
        random.shuffle(arr)
        train_files.extend(arr[:size])
    else:
        train_files.extend(arr)

for a in val_artists:
    arr = songs[a]

    if len(arr) > size:
        random.shuffle(arr)
        val_files.extend(arr[:size])
    else:
        val_files.extend(arr)

for a in test_artists:
    arr = songs[a]

    if len(arr) > size:
        random.shuffle(arr)
        test_files.extend(arr[:size])
    else:
        test_files.extend(arr)

print('train songs: {}, val songs: {}, test songs: {}'.format(len(train_files), len(val_files), len(test_files)))

dest = 'data/train/'
for f in tqdm(train_files):
    os.rename(parent_folder + f, dest + f)

dest = 'data/validate/'
for f in tqdm(val_files):
    os.rename(parent_folder + f, dest + f)

dest = 'data/test/'
for f in tqdm(test_files):
    os.rename(parent_folder + f, dest + f)


```
